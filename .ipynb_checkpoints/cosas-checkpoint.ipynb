```python
import tensorflow as tf

def scan_picture(path_to_picture):
    print(path_to_picture)
    img = tf.io.read_file(path_to_picture)
    print("---------------------")
    print(img)
    img = tf.image.decode_image(img)
    print("---------------------")
    print(img)
    img = tf.image.resize(img, size=[512, 384])
    print("---------------------")
    print(img)
    img = img/255.
    print("---------------------")
    print(img)
    return img




scan_picture("img/car.jpg")
```
